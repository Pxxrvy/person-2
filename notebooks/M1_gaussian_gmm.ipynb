```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/dataset.csv")
df.head()
```

```python
safe = df[df["label"] == 0].drop(columns=["label"]).values
hotspot = df[df["label"] == 1].drop(columns=["label"]).values
```

```python
print(safe.shape)
print(hotspot.shape)
```

```python
mu_safe = np.mean(safe, axis=0)
mu_hotspot = np.mean(hotspot, axis=0)

print("Safe mean:", mu_safe)
print("Hotspot mean:", mu_hotspot)
```

```python
sigma_safe = np.cov(safe, rowvar=False)
sigma_hotspot = np.cov(hotspot, rowvar=False)

print("Safe variance:", sigma_safe)
print("Hotspot variance:", sigma_hotspot)
```

```python
import pickle, os

os.makedirs("../results", exist_ok=True)

params = {
    "safe": {"mu": mu_safe, "sigma": sigma_safe},
    "hotspot": {"mu": mu_hotspot, "sigma": sigma_hotspot}
}

with open("../results/gaussian_params.pkl", "wb") as f:
    pickle.dump(params, f)
```
